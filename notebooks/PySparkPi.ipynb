# PySpark Pi on YARN
Вычисление числа Pi с использованием PySpark на YARN.


```python
from pyspark.sql import SparkSession
import random
```

```python
spark = SparkSession.builder.master('yarn').appName('PySparkNotebookPi').getOrCreate()
```

```python
n = 100000
rdd = spark.sparkContext.parallelize(range(n), 2)
inside = rdd.map(lambda _: 1 if (random.random()**2 + random.random()**2) <= 1 else 0).reduce(lambda a,b: a+b)
pi = 4.0 * inside / n
print(f'Pi is roughly {pi}')
```
