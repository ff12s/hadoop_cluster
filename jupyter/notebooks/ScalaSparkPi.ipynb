# Scala Spark Pi on YARN
Вычисление числа Pi с использованием Scala (Toree kernel) на YARN.


```scala
import org.apache.spark.sql.SparkSession
val spark = SparkSession.builder().master("yarn").appName("ScalaNotebookPi").getOrCreate()

val n = 100000
val rdd = spark.sparkContext.parallelize(1 to n, 2)
val inside = rdd.map { _ =>
  val x = scala.util.Random.nextDouble()
  val y = scala.util.Random.nextDouble()
  if (x*x + y*y <= 1.0) 1 else 0
}.reduce(_ + _)
val pi = 4.0 * inside / n
println(s"Pi is roughly $pi")

spark.stop()
```
